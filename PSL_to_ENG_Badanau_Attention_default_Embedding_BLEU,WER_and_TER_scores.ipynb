# Compute BLEU Score

```python
from nltk.translate.bleu_score import sentence_bleu
```

```python
i=0
gram1=0
while i<len(predicted):
  print(f'Predicted-> {predicted[i]}')
  print(f'Actual-> {groundtruth[i]}')
  print(f'1-Gram-> {sentence_bleu([groundtruth[i].split()],predicted[i].split(),weights=(1.0, 0, 0, 0))}')
  gram1+=sentence_bleu([groundtruth[i].split()],predicted[i].split(),weights=(1.0, 0, 0, 0))
  i=i+1
  print()
```

```python
gram1/len(predicted)
```

# WER

```python
from jiwer import wer
```

```python
i=0
sum=0
while i <len(groundtruth):
  print(f'Groundtruth----> {groundtruth[i]}')
  print(f'Predicted---> {predicted[i]}')
  print(f'Score {wer(groundtruth[i],predicted[i])}')
  sum+=wer(groundtruth[i],predicted[i])
  i+=1
```

```python
sum/len(groundtruth)
```
